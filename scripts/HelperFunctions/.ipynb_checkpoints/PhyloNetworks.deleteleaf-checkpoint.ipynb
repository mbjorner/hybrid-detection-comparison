The following is testing the functionality of PhyloNetworks' [deleteleaf! function](http://crsl4.github.io/PhyloNetworks.jl/latest/lib/public/#PhyloNetworks.deleteleaf!), to see if it can correctly identify hybrid nodes, and removes nodes as appropriate.

its arguments are

simplify - removes hybrids in some cases (if "forming a cycle of k=2 nodes, then these hybrid edges are merged and simplified as a single tree edge.")

nofuse - forces simplify to false

keeporiginalroot 

multigammas - ?

unroot - should not be used, we are not trying to unroot the network

```julia
import Pkg; Pkg.add("PhyloNetworks")
import Pkg; Pkg.add("PhyloPlots")

using PhyloPlots
using PhyloNetworks
```

Results: 

- simplify as "true" removes some hybrid clades, 
- nofuse as false is the same as nofuse unspecified,
- setting nofuse to true both forces simplify to false and produces odd graphs.


nofuse - true for the following: (forces simplify to false)
- conditions 1 & 2 match (root - false)
- conditions 3 & 4 match (root - true)

nofuse - false for the following
- cond 5 unique (root false, simplify true)
- cond 6 unique (root false, simplify false)
- cond 7 unique (root true, simplify true)
- cond 8 unique (root true, simplify false)

whether or not keep root is set as true or false, this does not impact the number or identity of hybrids, however - simplify as true gives a different (fewer) number of hybrids



Impact: keep nofuse undefined, rooting doesn't matter: but should simplify be used?

RE: "ghost lineages" are there when simplify is false ?
